## The change in the number of 100% free meals schools.

To achieve this, let me sort data according to free meals applying rate.

```python
data_1718_clean.sort_values(by=['1718_Free_Meal_Percent'], ascending=False)
```

It's not that helpful. Let me filter out schools with a Free Meals Percentage of 100%.

```python
data_1718_clean[data_1718_clean['1718_Free_Meal_Percent'] == 1.0]
```

There are 175 schools, but many schools are juvenile court schools. I'd like to see data to exclude juvenile court schools. (Because I'd like to see the change of 1819 and 2122, but juvenile court schools' free meals rate would stay at 100%, so...)

```python
all_free_meals_clean_1718 = data_1718_clean[
    (data_1718_clean['1718_Free_Meal_Percent'] == 1.0) &
    (data_1718_clean['School Type'] != 'Juvenile Court Schools')
].copy()
all_free_meals_clean_1718
```

Now we can see there're 110 schools for the year 17-18. (By the way, I'm surprised by the number of Juvenile Court Schools...)

Then how about other years?

```python
all_free_meals_clean_1819 = data_1819_clean[
    (data_1819_clean['1819_Free_Meal_Percent'] == 1.0) &
    (data_1819_clean['School Type'] != 'Juvenile Court Schools')
].copy()
all_free_meals_clean_1819
```

```python
all_free_meals_clean_1920 = data_1920_clean[
    (data_1920_clean['1920_Free_Meal_Percent'] == 1.0) &
    (data_1920_clean['School Type'] != 'Juvenile Court Schools')
].copy()
all_free_meals_clean_1920
```

```python
all_free_meals_clean_2021 = data_2021_clean[
    (data_2021_clean['2021_Free_Meal_Percent'] == 1.0) &
    (data_2021_clean['School Type'] != 'Juvenile Court Schools')
].copy()
all_free_meals_clean_2021
```

```python
all_free_meals_clean_2122 = data_2122_clean[
    (data_2122_clean['2122_Free_Meal_Percent'] == 1.0) &
    (data_2122_clean['School Type'] != 'Juvenile Court Schools')
].copy()
all_free_meals_clean_2122
```

This is interesting, the data from 1718 to 2122 is: 110, 118, 104, 121, 110. Also noted that the enrollment in 2122 is far lower than 1718, so it's safe to say:

4. After the pandemic, proportion of schools with 100% free meals is greater than before the pandemic.
