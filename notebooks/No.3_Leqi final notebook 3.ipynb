## LA School free meal count

I'm filtering out LA's schools data.

```python
LA_schools_2122 = data_2122_clean[
    (data_2122_clean['County Name'] == 'Los Angeles')
].copy()
LA_schools_2122
```

```python
alt.Chart(LA_schools_2122).mark_bar().encode(
    x='District Name',
    y='2122_Free_Meal_Percent'
).properties(
    title={
        'text': 'Los Angeles Schools',
        'subtitle' : ['Percentage of free meals by school districts']
    }
)
```

## LA School free meals by type of School

```python
alt.Chart(LA_schools_2122).mark_bar().encode(
    x='School Type',
    y='2122_Free_Meal_Percent'
).properties(
    title={
        'text': 'Los Angeles Schools',
        'subtitle' : ['Percentage of free meals by school type']
    }
)
```

## LA traditional high school free meal percent

First, let's pull all the traditional school data.

```python
traditional_schools_2122 = data_2122_clean[
    (data_2122_clean['Educational_Type'] == 'Traditional')
].copy()
traditional_schools_2122
```

```python
traditional_schools_2122.sort_values(by=['2122_Free_Meal_Percent'], ascending=False)
```

There're 8854 rows, let's narrow it down a bit.

```python
traditional_schools_2122['School Type'].value_counts()
```

Okay, I would like to make charts about middle and high schools.

```python
traditional_high_schools_2122 = traditional_schools_2122[
    (traditional_schools_2122['School Type'] == 'High Schools (Public)') |
    (traditional_schools_2122['School Type'] == 'Junior High Schools (Public)') |
    (traditional_schools_2122['School Type'] == 'Intermediate/Middle Schools (Public)')
].copy()
traditional_high_schools_2122
```

```python
alt.Chart(traditional_high_schools_2122).mark_bar().encode(
    x='County Name',
    y='2122_Free_Meal_Percent'
).properties(
    title={
        'text': 'Traditional high schools',
        'subtitle' : ['Percentage of free meals by county']
    }
)
```
